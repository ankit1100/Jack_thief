```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

per_player_cards=10 #no of cards per player
no_of_players=8 #no of player
no_of_cards=(per_player_cards*no_of_players)+1

card_list=[]
card_list.append(11) #jack is given card number 11
#no of cards-1=81-1=80 cards
for i in range(no_of_cards-1):
  card_list.append(int((i/4)+1)) #every number has 4 copies
#print(card_list)

count_jacks=0
#count the number of jacks since we need to see if the number of jacks are odd or even
for i in range(len(card_list)):
  if(card_list[i]==11):
    count_jacks=count_jacks+1

#if the number of jacks are even, then we will remove 1 jack from the entire set
if(count_jacks%2==0):
  card_list.pop(0)
  
#print(card_list)

starter=np.random.uniform(0,no_of_players)
st=int(starter)

player=[]
#storing cards of all the  players
for i in range(no_of_players):
  card_of_player=[] #cards of the particular player are stored here
  if(i==0): #give one extra cards to the first player
    for j in range(per_player_cards+1):
      rndnum=len(card_list)*np.random.uniform(0,1)
      rdnum=int(rndnum)
      card_of_player.append(card_list[rdnum])
      card_list.pop(rdnum)
  else: #do not give one extra cards to the other players
    for j in range(per_player_cards):
      rndnum=np.random.uniform(0,len(card_list))
      rdnum=int(rndnum)
      card_of_player.append(card_list[rdnum])
      card_list.pop(rdnum)
  #append the values in the end
  player.append(card_of_player)
  
#print(player)

for i in range(no_of_players):
  cards_paired=0
  for j in range(len(player[i])):
    for k in range(j):
      #assign mathcing cards 0 value
      if(player[i][j]==player[i][k]):
        player[i][j]=0
        player[i][k]=0
        cards_paired=cards_paired+2
  cards_left=len(player[i])
  for j in range(cards_paired):
    for k in range(cards_left):
      if(player[i][k]==0):
        #pop the value of the cards whose value os zero
        player[i].pop(k)
        break
    cards_left=cards_left-1
  
#print(player)
   
no_of_plays=0
while(True):
  for i in range(no_of_players):
      '''
      marked_positions=[]
      for j in range(len(player[i])):
        for k in range(j):
          if(player[i][j]==player[i][k]):
            break
      '''
      if(i==no_of_players-1):#passing of cards from 1 player to the other player
        rndnum=len(player[i])*np.random.uniform(0,1)
        rdnum=int(rndnum)
        player[0].append(player[i][rdnum])
        player[i].pop(rdnum)
      else: #other condition of passing of cards from one player to the other
        rndnum=len(player[i])*np.random.uniform(0,1)
        rdnum=int(rndnum)
        player[i+1].append(player[i][rdnum])
        player[i].pop(rdnum)
      cards_paired=0
      
      if(i==no_of_players-1):
        nm=0
        passing=False
        #check if the next player has cards with him or her or not
        while(True):
          if(len(player[nm])==0):
            if(nm==no_of_players-1):
              nm=0
            else:
              nm=nm+1
          else:
            passing=True
            break
        for j in range(len(player[nm])):
          for k in range(j):
            if(player[nm][j]==player[nm][k]):
              player[nm][j]=0
              player[nm][k]=0
              cards_paired=cards_paired+2
        cards_left=len(player[nm])
        for j in range(cards_paired):
          for k in range(cards_left):
            if(player[nm][k]==0):
              player[nm].pop(k)
              break
          cards_left=cards_left-1
      
      else:
        nm=i+1
        passing=False
        while(True):
          if(len(player[nm])==0):
            if(nm==no_of_players-1):
              nm=0
            else:
              nm=nm+1
          else:
            passing=True
            break
        for j in range(len(player[nm])):
          for k in range(j):
            if(player[nm][j]==player[nm][k]):
              player[nm][j]=0
              player[nm][k]=0
              cards_paired=cards_paired+2
        cards_left=len(player[nm])
        for j in range(cards_paired):
          for k in range(cards_left):
            if(player[nm][k]==0):
              player[nm].pop(k)
              break
          cards_left=cards_left-1    
  
  no_of_plays=no_of_plays+1
  print(player)
  iscontinue=False
  
  for i in range(no_of_players):
    if(len(player[i])>1):
      iscontinue=True
  if(iscontinue==False):
    break
  
  
print(no_of_plays)
#print(player)
      
     
  
```
